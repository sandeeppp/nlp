```python
import pandas as pd
import re
from collections import defaultdict
import numpy as np
import math
```

```python
def tokenize(msg):
    msg=msg.lower()
    wods=re.findall("[a-z0-9']+",msg)
    return(set(wods))
k=data.iloc[:4000].values
counts=defaultdict(lambda:[0,0])
def count_words(k):
    counts=0
    counts=defaultdict(lambda:[0,0])
    for i in k:
        for word in tokenize(i[1]):
            counts[word][0 if i[0]=='ham' else 1]+=1
    return counts

l=count_words(k)
gg=[]
def word_probs(l):
    gg=[]
    for i ,j in l.items():
        gg.append([i, (0.5+j[0])/(j[0]+j[1]+1), (0.5+j[1])/(j[0]+j[1]+1)])
    return gg
ggg=word_probs(l)
def spam_probs(j):
    a=tokenize(j)
    logspamprob=0
    loghamprob=0
    for word,spamprob,hamprob in ggg:
        if word in a:
            loghamprob+=math.log(hamprob)
            logspamprob+=math.log(spamprob)
    return ['spam' if loghamprob>logspamprob else 'ham']  
#this last step is something to think about 
```

```python
testdata=data.iloc[4000:].values
testy=[]
testx=[]
for i,j in testdata:
    testy.append(i)
    testx.append(j)
kk=[]
for d in testx:
    kk.append(spam_probs(d))
    

    
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(testy,kk)
```
