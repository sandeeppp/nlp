```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
iris.data
iris.target
iris.feature_names
#iris.target_names
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler
from sklearn.linear_model import LogisticRegression
x=iris.data[:,[1,2]]
y=iris.target
```

```python
lr=LogisticRegression()
from sklearn.model_selection import train_test_split
trainx,testx,trainy,testy=train_test_split(x,y)
```

```python
lr.fit(trainx,trainy)
pre=lr.predict(trainx)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
confusion_matrix(pre,testy)
```

```python
lr.score(trainx,trainy)
```

```python
lr.score(testx,testy)
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
xx1=np.arange(trainx[:,0].min(),trainx[:,0].max(),0.02)#shape is (110,)
xx2=np.arange(trainx[:,1].min(),trainx[:,1].max(),0.02)#shape is (295,)
c,d=np.meshgrid(xx1,xx2)# both shapes are (295,110)
z=lr.predict(np.array([c.ravel(),d.ravel()]).T)# shape of c.ravel,b.ravel is (32450,)........the whole input of lrpredict is (32450,2).....and the output is (32450)
z=z.reshape(c.shape)
plt.contourf(c,d,z, cmap='jet')# z should be in the shape of c and d and should contain the answers
plt.colorbar() 
k=zip(trainx.tolist(),lr.predict(trainx))
plt.scatter(trainx[pre==0,0],trainx[pre==0,1])
plt.scatter(trainx[pre==1,0],trainx[pre==1,1])
plt.scatter(trainx[pre==2,0],trainx[pre==2,1])
plt.show()
```
